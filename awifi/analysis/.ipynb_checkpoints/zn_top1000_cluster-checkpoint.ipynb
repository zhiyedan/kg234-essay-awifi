```python
import pandas as pd
import time
import datetime
import numpy as np
import split_data
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl  
import warnings
warnings.filterwarnings('ignore')
import plot_show
import tslearn_cluster
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data/zn_top1000_24h.csv",sep=',')
df.head()
```

```python
scaler = MinMaxScaler()
scaler_data = scaler.fit_transform(df)
scaler_data.shape
```

```python
center_num=5
cluster = tslearn_cluster.TsCluster(df,center_num=center_num)
cluster.run()
```

```python
# one_label = cluster.get_one_label_index(4)
# plot_show.plt_random_cols(df=df,cols_index=one_label,rand_col_num=5)
```

```python
one_label = cluster.get_one_label_index(4)
plot_show.plt_random_numpy(scaler_data=scaler_data,cols_index=one_label,rand_col_num=20)
```

```python
def plot_all_scaled_type():
    all_data = []
    for i in range(center_num):
        lable_index = cluster.get_one_label_index(i)
        one_class_data = scaler_data[:,lable_index]
        mean_24 = one_class_data.mean(axis=1)
        all_data.append(mean_24)
    plot_show.plt_one_ts(all_data)
plot_all_scaled_type()
```
