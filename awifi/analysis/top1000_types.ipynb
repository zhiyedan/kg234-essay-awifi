```python
import pandas as pd
import time
import datetime
import random
import numpy as np
import split_data
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl  
import warnings
warnings.filterwarnings('ignore')
import plot_show
import tslearn_cluster
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data/zn_top1000_24h.csv",sep=',')
df.head()
```

```python
cols = df.columns.values
```

```python
scaler = MinMaxScaler()
scaler_data = scaler.fit_transform(df)
```

```python
types_df = pd.read_csv("../data/eqp_type.txt",sep='\t',names=['ap_mac','company','address','type'])
types_df.head(10)
```

```python
type_result = []
for col in cols:
    t = types_df[types_df.ap_mac==col]['type'].values
    result = 0 if len(t)==0 else t[0]
    type_result.append(result)
```

```python
type_100 = []
for i in range(len(cols)):
    col = cols[i]
    t = types_df[types_df.ap_mac==col]
    if len(t)!=0:
        if t['type'].values[0]==100:
            type_100.append(i)
#             print(t['company'].values,t['address'].values)
```

```python
from collections import defaultdict
count_dict = defaultdict(int)
for x in type_result:
    count_dict[x] += 1
print(count_dict)
```

```python
this_type = 0
plot_num = 250
tt = np.array(type_result)
type_2 = np.where(tt==this_type)
# todo 真是麻烦。。。
col_indexs = list(type_2[0])
type_2_num = count_dict[this_type]
rand_col_num = plot_num if type_2_num>=plot_num else type_2_num
plot_show.plt_random_numpy(scaler_data=scaler_data,cols_index=col_indexs,rand_col_num=rand_col_num)
```
